```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("anime.csv")
data.head()
```

```python
data.info()
```

```python
data['type'].unique()
```

```python
Rreaview = 'TV'
data[data['type']==Rreaview]
```

```python
unique_countries = data[data.members>6000]['name'].unique()
unique_countries.sort()
unique_countries
```

```python
country_confirmed_case = []
for i in unique_countries:
  country_confirmed_case.append(data[data.members>6000][data['name']==i][data['type']==Rreaview].members.sum())
  
```

```python
print(set(zip(unique_countries, country_confirmed_case)))
```

```python
unique_provinces = data['episodes'][data.members>6000].unique()
unique_provinces
```

```python
country_confirmed_case = []
for i in unique_countries:
  country_confirmed_case.append(data[data.members>0][data['episodes']==i][data['type']==Rreaview].members.sum())
```

```python
print(set(zip(unique_countries, country_confirmed_case)))
```

```python
unique_provinces = data['episodes'][data.members>6000].unique()
unique_provinces
```

```python
province_confirmed_case = []
for i in unique_provinces:
  province_confirmed_case.append(data[data.members>6000][data['episodes']==i][data['type']==Rreaview].members.sum())
```

```python
print(set(zip(unique_provinces, province_confirmed_case)))
```

```python
nan_indices = []
for i in range(len(unique_provinces)):
  if type(unique_provinces[i])==float:
    nan_indices.append(i)
```

```python
unique_provinces = list(unique_provinces)
province_confirmed_case = list(province_confirmed_case)
```

```python
for i in nan_indices:
  unique_provinces.pop(i)
  province_confirmed_case.pop(i)

unique_provinces
```

```python
plt.barh(unique_countries, country_confirmed_case)
plt.show()
```

```python
plt.figure(figsize=(50,40))
plt.barh(unique_provinces, province_confirmed_case)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(country_confirmed_case)
plt.legend(unique_countries, loc='best')
plt.show()
```
